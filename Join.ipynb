```python
import pandas as pd 
```

```python
prod_cat_name = pd.read_csv('product_category_name_translation.csv')
sel_data7 = pd.read_csv('olist_sellers_dataset.csv')
prod_data6 = pd.read_csv('olist_products_dataset.csv')
order_data2 = pd.read_csv('olist_orders_dataset.csv')
order_rev1 = pd.read_csv('olist_order_reviews_dataset.csv')
order_pay3 = pd.read_csv('olist_order_payments_dataset.csv')
order_it_data5 = pd.read_csv('olist_order_items_dataset.csv')
geo8 = pd.read_csv('olist_geolocation_dataset.csv') # --> Not to merge
cust_data4 = pd.read_csv('olist_customers_dataset.csv')
```

```python
Merge1 = order_rev1.merge(order_data2, on='order_id', how='left', indicator=False)
Merge2 = Merge1.merge(order_pay3, on='order_id', how='left', indicator=False)
Merge3 = Merge2.merge(cust_data4, on='customer_id', how='left', indicator=False)
Merge4 = Merge3.merge(order_it_data5, on='order_id', how='left', indicator=False)
Merge5 = Merge4.merge(prod_data6, on='product_id', how='left', indicator=False)
Merge6 = Merge5.merge(sel_data7, on='seller_id', how='left', indicator=False)
```
